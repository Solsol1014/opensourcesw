```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
files = ['koreaAll2022.csv', 'Seoul2022.csv', 'Daejun2022.csv', 'Busan2022.csv', 'Jeju2022.csv']
#files = ['koreaAll2022.csv']

w = 0.34
x = np.arange(1, 13)
allav = np.array([])

for file in files:
    f = open(file, 'r', encoding='cp949')
    data = csv.reader(f)

    for i in range(8):
        h = next(data)

    y = []

    for row in data:
        if len(row)<5:
            break

        y.append(float(row[2]))
        allav = np.append(allav, float(row[2]))

    plt.bar(x-w, y, label=file.split('.')[0], width=0.17)
    w-=0.17
    f.close()

plt.title('Average monthly temperature by region')
plt.xticks(np.arange(1, 13))
plt.yticks(np.arange(-5, 31, 5))
plt.legend()
plt.show()
```

```python
w = 0.4
for file in files:
    if file=='koreaAll2022.csv':
        pass
    else:
        f = open(file, 'r', encoding='cp949')
        data = csv.reader(f)

        for i in range(8):
            h = next(data)
        
        y = []
        
        i = 0
        for row in data:
            if len(row)<5:
                break

            y.append(float(row[2])-allav[i])
            i+=1

        plt.bar(x-w+0.1, y, label=file.split('.')[0], width=0.2)
        w-=0.2
        f.close()

plt.title('Comparison of national average temperature and regional temperature')
plt.xticks(np.arange(1, 13))
plt.yticks(np.arange(-2, 9, 1))
plt.legend()
plt.show()
```
